```python
df=pd.read_excel('제공데이터.xlsx', encoding='utf-8')
df.head(5)
```

```python
del df['Unnamed: 0']
del df['Unnamed: 0.1']
```

```python
df=df.drop(df[df['취급액']==0].index)
```

```python
x=df.iloc[:, [2,3, 5, 6, 8, 10, 11, 13, 16]]
y=df.iloc[:, [7]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

XGBoost

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

gb=GradientBoostingRegressor(learning_rate=0.1,max_depth=5, max_features=4, min_samples_leaf=3, min_samples_split=4, n_estimators=1000)
gb.fit(x_train, y_train)

gb_pred = gb.predict(x_test)

np.sqrt(mean_squared_error(y_test, gb_pred))
```

```python
gb.score(x_test, y_test)
```
